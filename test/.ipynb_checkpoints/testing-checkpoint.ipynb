```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open("D:/STUDY/Fouth Year/GP/test/Ibrahim/Ibrahim_L_Index-finger.bmp")
print(type(img))
print(img)

# summarize some details about the image
print(img.format)
print(img.size)
print(img.mode)
```

```python
img_np_array = np.array(img)
print(type(img_np_array))
print(img_np_array.shape)
```

```python
rgb_array = np.delete(arr=img_np_array, obj=3, axis=2)
print(rgb_array.shape)
```

```python
from PIL import Image
import cv2
import matplotlib.pyplot as plt
resized_photo = cv2.resize(rgb_array, (90, 90))
```

```python
real_photo = resized_photo.astype(np.float32) / 255.
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.imshow(real_photo.squeeze(),cmap='gray')
```

```python
with open('numpy_photo.npz', 'wb') as logo_file:
  np.save(logo_file, real_photo)
```

```python
with open('numpy_photo.npz', 'rb') as logo_file:
  logo_array = np.load(logo_file)
print(logo_array.shape)
plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.imshow(logo_array)
```
