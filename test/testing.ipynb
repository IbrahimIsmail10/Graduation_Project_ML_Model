```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = Image.open("D:/STUDY/Fouth Year/GP/test/Ibrahim/100__M_L_Index_finger.bmp")
print(type(img))
print(img)

# summarize some details about the image
print(img.format)
print(img.size)
print(img.mode)
```

```python
img_np_array = np.array(img)
print(img_np_array.shape)
print(img.mode)
```

```python
rgb_array = np.delete(arr=img_np_array, obj=3, axis=2)
print(rgb_array.shape)
print(img.mode)
```

```python
from PIL import Image
import cv2
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
resized_photo = cv2.resize(img_np_array, (90, 90))
```

```python
gray_image = rgb2gray(resized_photo)

# Add a new axis to make the shape (90, 90, 1)
gray_image = np.expand_dims(gray_image, axis=-1)
```

```python
real_photo = gray_image.astype(np.float32) / 255.
```

```python
print(real_photo.shape)
```

```python
plt.figure(figsize=(8, 4))
plt.subplot(1, 3, 1)
plt.imshow(real_photo.squeeze(),cmap='gray')
```

```python
with open('numpy_photo.npz', 'wb') as logo_file:
  np.save(logo_file, real_photo)
```
